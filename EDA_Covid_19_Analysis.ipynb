```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```
data=pd.read_csv('covid_19_data (1).csv')
```

```
## Print top 5 rows of the data
data.head()
```

```
## Print last 5 rows of data
data.tail()
```

```
## Check for the null values and the datatype in the columns
data.info()
```

```
data['ObservationDate']=pd.to_datetime(data['ObservationDate'])
```

```
data.info()
```

```
## Rename the columns
data=data.rename(columns={"ObservationDate":"date","Country/Region":"country"})
```

```
data
```

```
## Find out the number of active cases
data['Active']=data['Confirmed']-data['Deaths']-data['Recovered']
data.tail()
```

```
## Get the latetst  max numbers according to the date
latest=data[data['date']==data['date'].max()]
latest.head()
```

```
len(latest)
```

```
latest['country'].value_counts()
```

```
world=latest.groupby("country")["Confirmed","Deaths","Recovered","Active"].sum()
world=world.reset_index()
world.head()
```

```
world['country'].value_counts()
```

```
## world map plot
world_map=px.choropleth(world,locations="country",locationmode="country names",
                        color_continuous_scale="greens",color="Recovered")
world_map.show()
```

```
world_map=px.choropleth(world,locations="country",locationmode="country names",
                        color_continuous_scale="reds",color="Deaths")
world_map.show()
```

```
world_map=px.choropleth(world,locations="country",locationmode="country names",
                        color_continuous_scale="blues",color="Confirmed")
world_map.show()
```

```
world_map=px.choropleth(world,locations="country",locationmode="country names",
                        color_continuous_scale="greys",color="Active")
world_map.show()
```

```
###Plot Active cases Vs Date graph
total_active_cases=data.groupby("date")["Active"].sum().reset_index()
total_active_cases
```

```
plt.figure(figsize=(50,20))
plt.xlabel("Days/date",fontsize=50)
plt.ylabel("Active cases",fontsize=50)
sns.pointplot(x=total_active_cases.index,y=total_active_cases['Active'])
plt.show()
```

```
total_active_cases=data.groupby("date")["Active","Confirmed","Recovered","Deaths"].sum().reset_index()
total_active_cases
```

```
plt.figure(figsize=(50,20))
plt.xlabel("Days/date",fontsize=50)
plt.ylabel("cases",fontsize=50)
sns.pointplot(x=total_active_cases.index,y=total_active_cases['Active'],color="blue")
sns.pointplot(x=total_active_cases.index,y=total_active_cases['Confirmed'],color="orange")
sns.pointplot(x=total_active_cases.index,y=total_active_cases['Recovered'],color="green")
sns.pointplot(x=total_active_cases.index,y=total_active_cases['Deaths'],color="red")
plt.show()
```

```
### Top 20 countries with maximum confirmed cases and check the recovery dates
top=latest.groupby("country")['Confirmed','Recovered'].sum()
top=top.sort_values("Confirmed",ascending=False).reset_index()
top_20=top.head(20)
top_20
```

```
## Bar plot of confirmed cases and recovered cases per country wise
plt.figure(figsize=(12,6))
sns.barplot(x=top_20['Confirmed'],y=top_20['country'],color="red",label="Confirmed cases")
sns.barplot(x=top_20['Recovered'],y=top_20['country'],color="green",label="Recovered cases")
plt.title("Recovery rate",fontsize=20)
plt.xlabel("cases",fontsize=15)
plt.ylabel("countries",fontsize=15)
plt.legend(loc='best')
plt.show()
```
